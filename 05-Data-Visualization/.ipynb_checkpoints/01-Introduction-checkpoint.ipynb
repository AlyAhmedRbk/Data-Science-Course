```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
```

```python
# Adding labels in x and y axis
plt.plot(years, runs)
plt.xlabel('years')
plt.ylabel('runs')
plt.title("Sachin Tendulker's Yearly Runs ")
```

```python
# Multiple Lines in One Plot
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label='Virat Kohli')
plt.plot(years, sehwag, label='Virander Sehwag')


plt.xlabel('Years')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison')
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli)
plt.plot(years, sehwag)

plt.legend(['Kholi', 'Sehwag'])

plt.xlabel('Years')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison')
```

```python
plt.plot(years, kohli, 'ro--', label='Virat Kohli')
plt.plot(years, sehwag, 'g^:',label='Virander Sehwag')


plt.xlabel('Years')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison')
```

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")

plt.xlabel('Years')
plt.ylabel('Runs')
```

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
